The aim of this project is to clean the data and analyze the included used car listings.

The data dictionary provided with data is as follows:

* dateCrawled - When this ad was first crawled. All field-values are taken from this date.
* name - Name of the car.
* seller - Whether the seller is private or a dealer.
* offerType - The type of listing

* kilometer - How many kilometers the car has driven.
* monthOfRegistration - The month in which the car was first registered.
* fuelType - What type of fuel the car uses.
* brand - The brand of the car.
* notRepairedDamage - If the car has a damage which is not yet repaired.
* dateCreated - The date on which the eBay listing was created.
* nrOfPictures - The number of pictures in the ad.
* postalCode - The postal code for the location of the vehicle.
* lastSeenOnline - When the crawler saw this ad last online.

Modifying Column Names:

```python
column_rename_dict={
    'yearOfRegistration' : 'registration_year',
    'monthOfRegistration' : 'registration_month',
    'notRepairedDamage' : 'unrepaired_damage',
    'dateCreated' : 'ad_created',
}

def camelcase_to_snakecase(str):
    res = [str[0].lower()]
    for c in str[1:]:
        if c in ('ABCDEFGHIJKLMNOPQRSTUVWXYZ'):
            res.append('_')
            res.append(c.lower())
        else:
            res.append(c)
     
    return ''.join(res)

autos.rename(columns=column_rename_dict,inplace=True)
autos.rename(mapper=camelcase_to_snakecase,axis='columns',inplace=True)
#checking to see the changes:
autos.head()
```

Changes made:
* The column names use camelcase instead of Python's preferred snakecase, so those columns were converted to be in snakecase.
* Some column names were not very accurately describing what they represent, so those columns had their names modified based on the data dictionary provided.

When working with data on cars, it's natural to explore variations across different car brands. We will use aggregation to understand the brand column.

Analyzing whether the choosing top 20 car brands by sales volumes is good for further analysis or if selecting those car brands that account for over a certain percentage of the total car sales (e.g. > 5%) makes sense for proceeding further.:-

```python
brand_distribution=autos["brand"].value_counts()
print(brand_distribution)
```

Based on frequencies, it looks like 9 car brands (that have over a 1000 of its cars sold at Ebay), are in contention for further analysis.
Lets look at sale numbers in percentages.

```python
normalized_brand_distribution=autos["brand"].value_counts(normalize=True)*100
print(normalized_brand_distribution)
```

From this, clearly a 5% sales criteria would serve us better when we perform further analysis on car brands.

```python
five_pct_car_brands=normalized_brand_distribution.loc[normalized_brand_distribution>5.0].index
print(five_pct_car_brands)
```

Let us now aggregate on these brands to find out the mean price of car sales of these brands on Ebay

```python
brands_mean_car_price={}
for car_brand in five_pct_car_brands:
    brand_listings=autos.loc[autos["brand"]==car_brand]
    mean_brand_price=brand_listings["price"].mean()
    brands_mean_car_price[car_brand]=mean_brand_price

message_template="An {car_brand} car is sold at an average price of {mean_price} euros."

for brand, avg_price in brands_mean_car_price.items():
    print(message_template.format(car_brand=brand,mean_price=avg_price))
```

A mercedes_benz car sells for a much higher average price by a mile when compared to the other 5 car brands.
In the overall picture though:
* Audi, BMW and Mercedes Benz are more expensive
* Ford and Opel are less expensive
* Volkswagen is in between

For the top 6 brands, let's use aggregation to understand the average mileage for those cars and if there's any visible link with mean price. While our natural instinct may be to display both aggregated series objects and visually compare them, this has a few limitations:
* it's difficult to compare more than two aggregate series objects if we want to extend to more columns
* we can't compare more than a few rows from each series object
* we can only sort by the index (brand name) of both series objects so we can easily make visual comparisons

Let us now aggregate on these brands to find out the mean mileage for each of these brands on Ebay

```python
brands_mean_car_mileage={}
for car_brand in five_pct_car_brands:
    brand_listings=autos.loc[autos["brand"]==car_brand]
    mean_brand_mileage=brand_listings["odometer_km"].mean()
    brands_mean_car_mileage[car_brand]=mean_brand_mileage

message_template="An {car_brand} car is has an average mileage of {mean_mlg}."

for brand, avg_mlg in brands_mean_car_mileage.items():
    print(message_template.format(car_brand=brand,mean_mlg=avg_mlg))
```

We will now combine the data from both series objects into a single dataframe (with a shared index) and display the dataframe directly.

```python
#making both the dictionaries into series objects:
bmcp_series=pd.Series(brands_mean_car_price)
print(bmcp_series)
bmcm_series=pd.Series(brands_mean_car_mileage)
print(bmcm_series)
```

```python
#making a dataframe containing both series objects:
brands_mean_price_mileage = pd.DataFrame(bmcp_series, columns=['mean_price'])
print(brands_mean_price_mileage)
brands_mean_price_mileage["mean_mileage"]=bmcm_series
print(brands_mean_price_mileage)
```

Splitting the odometer_km into groups, and using aggregation to see if average prices follows any patterns based on the mileage:

```python
print(autos["odometer_km"].describe())
print(autos["odometer_km"].value_counts().sort_index())
```

splitting the values into groups, each group having a width of 10,000 km and using aggregation to see if average prices follows any patterns based on the mileage.

```python
odometer_groups_avg_prices={}
odometer_milestones=list(range(0,160000,10000))
for i in np.arange(0,len(odometer_milestones)-1,1):
    odometer_range_listings=autos.loc[autos["odometer_km"].between(odometer_milestones[i],odometer_milestones[i+1])]
    odometer_range_mean_price=odometer_range_listings["price"].mean()
    odometer_groups_avg_prices[str(odometer_milestones[i])+"-"+str(odometer_milestones[i+1])+"km"]=odometer_range_mean_price
ogap_series=pd.Series(odometer_groups_avg_prices)
print(ogap_series)
ogap_series.describe()
```

Determining how much cheaper cars are with damage than their non-damaged counterparts:-

```python
autos["unrepaired_damage"].unique()
damaged_car_listings = autos.loc[autos["unrepaired_damage"] == 'Yes']
avg_damaged_car_price=damaged_car_listings["price"].mean()
non_damaged_car_listings = autos.loc[autos["unrepaired_damage"] == 'No']
avg_non_damaged_car_price=non_damaged_car_listings["price"].mean()
cost_difference_btw_undamaged_and_damaged=avg_non_damaged_car_price-avg_damaged_car_price
print("Cars with their damages not repaired are on an average",cost_difference_btw_undamaged_and_damaged,"euros cheaper than cars having their damages repaired")
```
